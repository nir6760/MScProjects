### Problem 1

The PCA visualization provides a two-dimensional representation of the data obtained by applying PCA. PCA aims to capture the maximum variance in the data by identifying the principal components, which are orthogonal directions of maximum variability. In this case, the first two principal components were used to visualize the data.

The PCA visualization shows groupings of data points based on their similarities. However, due to its linear nature, PCA has limitations in capturing complex non-linear relationships present in the data.

In contrast, the Autoencoder visualization also presents a two-dimensional representation of the data after applying the autoencoder. Autoencoders are neural networks designed to learn a compressed representation of the input data in their hidden layers while aiming to reconstruct the original data.

The Autoencoder visualization reveals distinct clusters and separations between different groups of data points compared to PCA, as clearly shown in the figures. These clusters can be easily distinguished visually or by using clustering algorithms such as K-means. Autoencoders excel at capturing non-linear relationships and complex patterns in the data, resulting in more expressive and separated clusters.

The Linear Autoencoder visualization showcases compressed representations for most of the digits, making it challenging or even impossible to separate them. This observation aligns with the findings from the PCA visualization, which exhibit a similar shape but Linear Autoencoder have slightly better results in this case. The difference in performance between the Linear Autoencoder and PCA can be attributed to the contrasting optimization approaches used. The Linear Autoencoder benefits from its iterative neural network optimization process, which allows for improved results compared to the closed solution offered by PCA. Both methods rely solely on linear transformations. However, linear autoencoders (and PCA algorithm) lack the non-linear activation functions found in regular autoencoders, which limits their capacity to capture complex patterns and non-linear relationships in the data. Furthermore, upon visual examination, distinctions can be observed between PCA and Linear Autoencoder regarding the scaling and rotation of the input data's neurons or components. Despite their similarities, these two methods generate different coding spaces. However, it is possible to identify certain linear transformations that relate these spaces.

Overall, the Autoencoder outperforms both PCA and Linear Autoencoder in terms of separating the digits in the two-dimensional space. While PCA and Linear Autoencoder exhibit compressed representations with some grouping, they fail to capture the complexity and non-linearity present in the data, which is a notable difference compared to the Autoencoder. Additionally, the linear nature of PCA and Linear Autoencoder results in similar visualizations.

It is also worth mentioning that certain digits, such as 0 and 1, are easier to separate compared to digits like 3 and 4. The Autoencoder handles the separation of 4 quite well but struggles with 3. In each approach, the digit 8 appears to be centrally located within the two-dimensional space compared to the other digits, as its form combines elements from various other digits.
